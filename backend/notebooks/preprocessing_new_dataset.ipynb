# Preprocessing: District-wise Crime Dataset
This notebook converts the wide-format dataset (crime types as columns) into a long-format dataset required for machine learning.


```
import pandas as pd
from pathlib import Path

def prepare_new_dataset(input_path, output_path):
    # Load new dataset
    df = pd.read_csv(input_path)
    
    # Rename columns to standard names
    df = df.rename(columns={
        'STATE/UT': 'State',
        'DISTRICT': 'District',
        'YEAR': 'Year'
    })
    
    # List of crime columns (all columns except State, District, Year)
    crime_cols = [col for col in df.columns if col not in ['State', 'District', 'Year']]
    
    # Melt the wide format to long format
    df_long = df.melt(
        id_vars=['State', 'District', 'Year'],
        value_vars=crime_cols,
        var_name='Crime Description',
        value_name='Crime Count'
    )
    
    # Basic cleaning
    df_long['State'] = df_long['State'].str.strip().str.title()
    df_long['District'] = df_long['District'].str.strip().str.title()
    df_long['Crime Description'] = df_long['Crime Description'].str.strip()
    
    # Remove aggregate districts if any (sometimes datasets have 'TOTAL' rows)
    df_long = df_long[~df_long['District'].isin(['Total', 'Total Districts'])]
    
    # Save the prepared data
    df_long.to_csv(output_path, index=False)
    print(f"Melted data saved to {output_path}")
    print(f"Total rows: {len(df_long)}")

if __name__ == "__main__":
    import os
    project_root = Path(os.getcwd()).resolve()
    input_path = project_root / "data" / "raw" / "District_wise_crimes.csv"
    output_path = project_root / "data" / "processed" / "crime_data_district_long.csv"
    prepare_new_dataset(input_path, output_path)

```
